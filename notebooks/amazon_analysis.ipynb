```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/amazon.csv")
```

```python
#check the first 5 rows

df.head()
```

```python
#check the cloumns

df.columns
```

```python
#Check info

df.info()
```

```python
# Data Cleaning

#checking missing values

df.isnull().sum()
```

```python
# remove missing rows

df.dropna(inplace=True)
```

```python
# Remove Duplicates

df.drop_duplicates(inplace=True)
```

```python
#Clean Discounted Price

df['discounted_price'] = df['discounted_price'].str.replace('₹', '')
df['discounted_price'] = df['discounted_price'].str.replace(',', '')
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
# Clean Actual Price

df['actual_price'] = df['actual_price'].str.replace('₹', '')
df['actual_price'] = df['actual_price'].str.replace(',', '')
df['actual_price'] = df['actual_price'].astype(float)
```

```python
#Clean Discount Percentage

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
# Clean Rating Count
df['rating_count'] = df['rating_count'].str.replace(',', '')
```

```python
# Convert to numeric

df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# Basic Analysis


#Total Products

print("Total Products:", len(df))


#Average Rating

print("Average Rating:", df['rating'].mean())

#Average Discount

print("Average Discount:", df['discount_percentage'].mean())
```

```python
# Top Categories

df['main_category'] = df['category'].apply(lambda x: x.split('|')[0])


#Count Categories

category_count = df['main_category'].value_counts()
print(category_count)


#Visualize Categories

plt.figure(figsize=(10,5))

sns.barplot(
    x=category_count.index,
    y=category_count.values
)

plt.xticks(rotation=90)

plt.title("Top Product Categories")

plt.show()
```

```python
# Highest Rated Products

top_rated = df.sort_values(by='rating', ascending=False)

top_rated[['product_name', 'rating']].head(10)
```

```python
# Most Reviewed Products

most_reviewed = df.sort_values(by='rating_count', ascending=False)

most_reviewed[['product_name', 'rating_count']].head(10)
```

```python
# Discount Analysis

top_discount = df.sort_values(
    by='discount_percentage',
    ascending=False
)

top_discount[['product_name', 'discount_percentage']].head(10)
```

```python
# Price Analysis

plt.figure(figsize=(8,5))

sns.histplot(df['discounted_price'])

plt.title("Discounted Price Distribution")

plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(8,5))

sns.heatmap(
    df[['discounted_price',
        'actual_price',
        'discount_percentage',
        'rating',
        'rating_count']].corr(),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```
